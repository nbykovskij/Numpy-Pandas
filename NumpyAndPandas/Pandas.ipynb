```python
import numpy as np
import pandas as pd
```

```python
df_movie_raw = pd.read_csv('Pandas data/movies.csv', index_col = 'movieId')
df_rating_raw = pd.read_csv('Pandas data/ratings.csv', index_col = 'movieId')
```

## Задание 1

```python
#Лепим фрейм
merges_frame = pd.concat([df_rating_raw, df_movie_raw], axis=1, join='inner')
#Фильтруем фрейм
df_five_rating = merges_frame.loc[merges_frame['rating'] == 5.0]
#Формирую сводную
df_fp_rating = pd.pivot_table(df_five_rating,  
                   values = ['rating'],
                   index = ['title'],
                   aggfunc=(len),    
                  )

#Принимает на вход сводню таблицу и выплевывает интерпритированный ответ
def get_best(df, 
             word_interp = False, 
             maximum = True):
    
    new_dict = {}
    _dict = df.to_dict('index')
    
    for element, v in _dict.items():
        for item in v.values():
            new_dict[element] = item
            
    if word_interp == True and maximum == True:
        return print(f'Тут {max(new_dict, key=new_dict.get)} больше всего, а именно {max(new_dict.values())}')
    elif word_interp == False and maximum == True:
        return max(new_dict, key=new_dict.get), max(new_dict.values())
    elif word_interp == False and maximum == False:
        return sum(new_dict.values())
    else:
        pass
```

```python
get_best(df = df_fp_rating, 
              word_interp = True)
```
