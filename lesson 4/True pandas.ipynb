## Задание 1

```python
def traffic_channel(row):
    if row['traffic_source']=='yandex':
        return 'organic'
    
    elif row['traffic_source']=='google':
        return 'organic'
    
    elif row['traffic_source']=='paid':
        if row['region']=='Russia':
            return 'ad'
        else:
            return 'other'
        
    elif row['traffic_source']=='email':
        if row['region']=='Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return row['traffic_source']
    
log['channel_traffic'] = log.apply(traffic_channel, axis = 1)
log.head()
```

## Задание 3

```python
ratings_count = rating_raw.groupby('userId').count().reset_index() #считаем кол-во оценок по юзеру
count_more_than_100 = ratings_count[ratings_count['rating']>100] #создаем новый фрейм с юзерами 1--+ оценок

rating_max = rating_raw.groupby('userId').max().reset_index() #максимальные значения по юзеру
rating_min = rating_raw.groupby('userId').min().reset_index() #минимальные значения по юзеру

rating_max_timestamp = rating_max[['userId','timestamp']] #макс таймстамп
rating_min_timestamp = rating_min[['userId','timestamp']] #мин таймстамп

rating_max_timestamp.rename(columns = {'timestamp':'timestamp_max'},
                            inplace = True)

rating_min_timestamp.rename(columns = {'timestamp':'timestamp_min'}, 
                            inplace = True)

avg_timestamp = count_more_than_100.merge(rating_max_timestamp, how='inner').merge(rating_min_timestamp, how='inner')
```

```python
def average(row):
    mean_ =(row['timestamp_max']-row['timestamp_min'])
    return mean_

avg_timestamp['LT'] = avg_timestamp.apply(average,
                                          axis=1)
avg_timestamp
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111,
                      112,
                      113, 
                      114,
                      115],
        'rzd_revenue': [1093, 
                        2810,
                        10283, 
                        5774, 
                        981]
    }
)
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 
                      114, 
                      115, 
                      116, 
                      117],
        'auto_revenue': [57483, 
                         83, 
                         912, 
                         4834,
                         98]
    }
)
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 
                      116, 
                      117, 
                      118],
        'air_revenue': [81, 
                        4, 
                        13,
                        173]
    }
)
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 
                      112, 
                      113, 
                      114, 
                      115, 
                      116, 
                      117, 
                      118],
        'address': ['Комсомольская 4', 
                    'Энтузиастов 8а', 
                    'Левобережная 1а',
                    'Мира 14', 
                    'ЗЖБИиДК 1', 
                    'Строителей 18',
                    'Панфиловская 33',
                    'Мастеркова 4']
    }
)
```

```python
#мерджим табблицы
table = client_base.merge(rzd, how='outer', on='client_id')
table = table.merge(auto, how='outer', on='client_id')
table = table.merge(air, how='outer', on='client_id')

#избавляемся от NaN
table.loc[table.rzd_revenue.isnull(), 'rzd_revenue']=0
table.loc[table.auto_revenue.isnull(), 'auto_revenue']=0
table.loc[table.air_revenue.isnull(), 'air_revenue']=0

#Немного приправим функциями (получаем тотал sum)
def sum_rev(row):
    return row['rzd_revenue'] + row['auto_revenue'] + row['air_revenue']

table['total_rev'] = table.apply(sum_rev, axis = 1)
```

```python
full_table = table.sort_values('total_rev', ascending=False) #формируем фуловую таблицу
```
